### outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
numeric_cols=['Age','School_Grade','Daily_Usage_Hours','Sleep_Hours','Academic_Performance','Social_Interactions','Exercise_Hours','Screen_Time_Before_Bed','Phone_Checks_Per_Day','Apps_Used_Daily','Time_on_Social_Media','Time_on_Gaming','Time_on_Education','Family_Communication','Weekend_Usage_Hours','Addiction_Level']
sns.set(style='ticks')


for col in numeric_cols:
    sns.boxplot(y=Smartphone_df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
    

```
